```python
import math

num = int(input("Enter a number: "))

# Prime numbers are natural numbers greater than 1
if num <= 1:
    is_prime = False
elif num == 2:
    is_prime = True
else:
    is_prime = True
    # Check for divisors from 2 up to the square root of num
    # We only need to check up to int(math.sqrt(num)) + 1
    # because if a number has a divisor greater than its square root,
    # it must also have a divisor smaller than its square root.
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            is_prime = False
            break # No need to check further, it's not prime

if is_prime:
    print(f"{num} is a Prime number")
else:
    print(f"{num} is not a Prime number")
```

```python
import random

secret_number = random.randint(1, 50) # Generate a random number between 1 and 50
attempts = 7

print("Welcome to the Number Guessing Game!")
print(f"I'm thinking of a number between 1 and 50. You have {attempts} attempts.")

for attempt_num in range(1, attempts + 1):
    try:
        guess = int(input(f"Attempt {attempt_num}/{attempts}: Enter your guess: "))

        if guess < 1 or guess > 50:
            print("Your guess is out of range (1-50). Please try again.")
            continue

        if guess < secret_number:
            print("Too low!")
        elif guess > secret_number:
            print("Too high!")
        else:
            print(f"Correct! You guessed the number {secret_number} in {attempt_num} attempts!")
            break
    except ValueError:
        print("Invalid input. Please enter a whole number.")

else:
    print(f"Sorry, you've run out of attempts! The secret number was {secret_number}.")
print("Thanks for playing!")
```

```python
def temperature_converter(temp, unit="C"):
  if unit == "C":
    fahrenheit = (temp * 9/5) + 32
    return fahrenheit

  elif unit == "F":
    celsius = (temp - 32) * 5/9
    return celsius
  else:
    return "Invalid unit. Use 'C' for Celsius or 'F' for Fahrenheit."

print(temperature_converter(60, "F"))
```

```python
import math

def is_prime(n):
  if n <= 1:
    return False  # Numbers less than or equal to 1 are not prime
  elif n == 2:
    return True   # 2 is the only even prime number
  elif n % 2 == 0:
    return False  # Other even numbers are not prime
  else:
    # Check for divisors from 3 up to the square root of n (only odd numbers)
    for i in range(3, int(math.sqrt(n)) + 1, 2):
      if n % i == 0:
        return False
    return True

print(f"Is 7 prime? {is_prime(7)}")      # Expected: True
print(f"Is 10 prime? {is_prime(10)}")     # Expected: False
print(f"Is 1 prime? {is_prime(1)}")       # Expected: False
print(f"Is 2 prime? {is_prime(2)}")       # Expected: True
print(f"Is 29 prime? {is_prime(29)}")     # Expected: True
print(f"Is 9 prime? {is_prime(9)}")      # Expected: False
```

```python
def fibonacci(n):
    a, b = 0, 1
    count = 0

    if n <= 0:
        print("Please enter a positive integer.")
    elif n == 1:
        print("Fibonacci sequence:")
        print(a)
    else:
        print("Fibonacci sequence:")
        while count < n:
            print(a, end=" ")
            nth = a + b
            a = b
            b = nth
            count += 1
        print()

print("--- Test Cases ---")
fibonacci(7)
fibonacci(1)
fibonacci(0)
fibonacci(10)
```

```python
def stats(numbers):
    if not numbers:
        return {"error": "List is empty, cannot calculate statistics"}

    total_sum = sum(numbers)
    count = len(numbers)
    average = total_sum / count
    maximum = max(numbers)
    minimum = min(numbers)

    return {
        "sum": total_sum,
        "average": average,
        "max": maximum,
        "min": minimum
    }

# Test cases
list1 = [10, 20, 30, 40, 50]
print(f"Stats for {list1}: {stats(list1)}")

list2 = [5, 1, 9, 3, 7]
print(f"Stats for {list2}: {stats(list2)}")

empty_list = []
print(f"Stats for {empty_list}: {stats(empty_list)}")
```

```python
def calculator(num1, num2, operation):
    if operation == "+":
        return num1 + num2
    elif operation == "-":
        return num1 - num2
    elif operation == "*":
        return num1 * num2
    elif operation == "/":
        if num2 == 0:
            return "Error: Cannot divide by zero!"
        return num1 / num2
    else:
        return "Error: Invalid operation. Please use '+', '-', '*', or '/'."

# Example usage:
print(f"10 + 5 = {calculator(10, 5, '+')}")
print(f"10 - 5 = {calculator(10, 5, '-')}")
print(f"10 * 5 = {calculator(10, 5, '*')}")
print(f"10 / 5 = {calculator(10, 5, '/')}")
print(f"10 / 0 = {calculator(10, 0, '/')}")
print(f"10 ^ 5 = {calculator(10, 5, '^')}")
```

```python
def add(x, y):    return x + y
def subtract(x, y): return x - y
def multiply(x, y): return x * y
def divide(x, y):
    if y == 0: return "Cannot divide by zero!"
    return x / y

def calculator():
    print("=== Smart Calculator ===")
    while True:
        print("\n1. Add\n2. Subtract\n3. Multiply\n4. Divide\n5. Exit")
        choice = input("Choose (1-5): ")

        if choice == "5":
            print("Goodbye!")
            break

        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        if choice == "1":
            print("Result =", add(num1, num2))
        elif choice == "2":
            print("Result =", subtract(num1, num2))
        elif choice == "3":
            print("Result =", multiply(num1, num2))
        elif choice == "4":
            print("Result =", divide(num1, num2))
        else:
            print("Invalid choice!")

calculator()
```

```python
import csv

# Define the headers and some sample data for 5 students
headers = ["Name", "Math", "Science", "English"]
students_marks = [
    ["Alice", 85, 90, 78],
    ["Bob", 72, 65, 80],
    ["Charlie", 95, 88, 92],
    ["David", 60, 70, 75],
    ["Eve", 88, 92, 85]
]

file_name = "marks.csv"

# Write the data to the CSV file
with open(file_name, "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(headers) # Write the header row
    writer.writerows(students_marks) # Write the student data rows

print(f"File '{file_name}' created successfully with student marks!")

# Read and print the content of the file to verify
print(f"\nContent of {file_name}:")
with open(file_name, "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
import csv

def save_data(filename, data):
    """
    Saves a list of lists as a CSV file.

    Args:
        filename (str): The name of the CSV file to create.
        data (list of list): A list of lists representing the data to save.
                             Each inner list is a row in the CSV.
    """
    try:
        with open(filename, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(data)
        print(f"Data successfully saved to '{filename}'.")
    except IOError as e:
        print(f"Error saving data to file: {e}")

# Example usage:
example_data = [
    ["Name", "Age", "City"],
    ["Alice", 30, "New York"],
    ["Bob", 24, "London"],
    ["Charlie", 35, "Paris"]
]

save_data("example.csv", example_data)

# You can also verify the content by reading it back
print("\nVerifying content of example.csv:")
with open("example.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
#Function that reads CSV and returns list of dict
import csv

def read_csv_as_dict(filename):
    """
    Reads a CSV file and returns its content as a list of dictionaries.
    Each dictionary represents a row, with column headers as keys.

    Args:
        filename (str): The name of the CSV file to read.

    Returns:
        list: A list of dictionaries, where each dictionary is a row.
              Returns an empty list if the file is not found or empty.
    """
    data_list = []
    try:
        with open(filename, 'r', newline='') as file:
            reader = csv.DictReader(file)
            for row in reader:
                data_list.append(row)
        print(f"Data successfully read from '{filename}'.")
        return data_list
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
        return []
    except Exception as e:
        print(f"An error occurred: {e}")
        return []

# Example usage with the previously created 'example.csv':
data_from_csv = read_csv_as_dict("example.csv")

if data_from_csv:
    print("\nContent as list of dictionaries:")
    for row_dict in data_from_csv:
        print(row_dict)
```

```python
def add_task(task_description):
    """Appends a new task to the todo.txt file."""
    with open("todo.txt", "a") as file:
        file.write(task_description + "\n")
    print(f"Task '{task_description}' added.")

def view_tasks():
    """Reads and prints all tasks from the todo.txt file."""
    try:
        with open("todo.txt", "r") as file:
            tasks = file.readlines()
        if not tasks:
            print("Your todo list is empty!")
        else:
            print("\n--- Your Todo List ---")
            for i, task in enumerate(tasks, 1):
                print(f"{i}. {task.strip()}")
            print("----------------------")
    except FileNotFoundError:
        print("No todo list found. Start by adding a task!")

def todo_manager():
    """Main function to manage the todo list."""
    print("=== Todo List Manager ===")
    while True:
        print("\n1. Add Task\n2. View Tasks\n3. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            task = input("Enter the task description: ")
            add_task(task)
        elif choice == '2':
            view_tasks()
        elif choice == '3':
            print("Exiting Todo List Manager. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

# Run the todo manager
todo_manager()
```

```python
#Personal Diary/journal App
import datetime

def write_diary():
    date = datetime.date.today()
    print(f"\nWriting diary for {date}")
    entry = input("How was your day? Write your thoughts:\n")

    with open("my_diary.txt", "a") as file:
        file.write(f"\n{'='*50}\n")
        file.write(f"Date: {date}\n")
        file.write(f"Entry:\n{entry}\n")
        file.write(f"{'='*50}\n")
    print("Entry saved successfully! 📖")

def read_diary():
    try:
        with open("my_diary.txt", "r") as file:
            print(file.read())
    except FileNotFoundError:
        print("No diary entries yet. Start writing!")

while True:
    print("\n=== PERSONAL DIARY ===")
    print("1. Write new entry")
    print("2. Read all entries")
    print("3. Exit")
    choice = input("Choose (1-3): ")

    if choice == "1":
        write_diary()
    elif choice == "2":
        read_diary()
    elif choice == "3":
        print("See you tomorrow! Keep learning AI/ML 💪")
        break
    else:
        print("Invalid choice!")
```
