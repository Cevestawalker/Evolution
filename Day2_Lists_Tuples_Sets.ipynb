```python
import math

num = int(input("Enter a number: "))

# Prime numbers are natural numbers greater than 1
if num <= 1:
    is_prime = False
elif num == 2:
    is_prime = True
else:
    is_prime = True
    # Check for divisors from 2 up to the square root of num
    # We only need to check up to int(math.sqrt(num)) + 1
    # because if a number has a divisor greater than its square root,
    # it must also have a divisor smaller than its square root.
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            is_prime = False
            break # No need to check further, it's not prime

if is_prime:
    print(f"{num} is a Prime number")
else:
    print(f"{num} is not a Prime number")
```

```python
import random

secret_number = random.randint(1, 50) # Generate a random number between 1 and 50
attempts = 7

print("Welcome to the Number Guessing Game!")
print(f"I'm thinking of a number between 1 and 50. You have {attempts} attempts.")

for attempt_num in range(1, attempts + 1):
    try:
        guess = int(input(f"Attempt {attempt_num}/{attempts}: Enter your guess: "))

        if guess < 1 or guess > 50:
            print("Your guess is out of range (1-50). Please try again.")
            continue

        if guess < secret_number:
            print("Too low!")
        elif guess > secret_number:
            print("Too high!")
        else:
            print(f"Correct! You guessed the number {secret_number} in {attempt_num} attempts!")
            break
    except ValueError:
        print("Invalid input. Please enter a whole number.")

else:
    print(f"Sorry, you've run out of attempts! The secret number was {secret_number}.")
print("Thanks for playing!")
```

```python
def temperature_converter(temp, unit="C"):
  if unit == "C":
    fahrenheit = (temp * 9/5) + 32
    return fahrenheit

  elif unit == "F":
    celsius = (temp - 32) * 5/9
    return celsius
  else:
    return "Invalid unit. Use 'C' for Celsius or 'F' for Fahrenheit."

print(temperature_converter(60, "F"))
```

```python
import math

def is_prime(n):
  if n <= 1:
    return False  # Numbers less than or equal to 1 are not prime
  elif n == 2:
    return True   # 2 is the only even prime number
  elif n % 2 == 0:
    return False  # Other even numbers are not prime
  else:
    # Check for divisors from 3 up to the square root of n (only odd numbers)
    for i in range(3, int(math.sqrt(n)) + 1, 2):
      if n % i == 0:
        return False
    return True

print(f"Is 7 prime? {is_prime(7)}")      # Expected: True
print(f"Is 10 prime? {is_prime(10)}")     # Expected: False
print(f"Is 1 prime? {is_prime(1)}")       # Expected: False
print(f"Is 2 prime? {is_prime(2)}")       # Expected: True
print(f"Is 29 prime? {is_prime(29)}")     # Expected: True
print(f"Is 9 prime? {is_prime(9)}")      # Expected: False
```

```python
def fibonacci(n):
    a, b = 0, 1
    count = 0

    if n <= 0:
        print("Please enter a positive integer.")
    elif n == 1:
        print("Fibonacci sequence:")
        print(a)
    else:
        print("Fibonacci sequence:")
        while count < n:
            print(a, end=" ")
            nth = a + b
            a = b
            b = nth
            count += 1
        print()

print("--- Test Cases ---")
fibonacci(7)
fibonacci(1)
fibonacci(0)
fibonacci(10)
```

```python
def stats(numbers):
    if not numbers:
        return {"error": "List is empty, cannot calculate statistics"}

    total_sum = sum(numbers)
    count = len(numbers)
    average = total_sum / count
    maximum = max(numbers)
    minimum = min(numbers)

    return {
        "sum": total_sum,
        "average": average,
        "max": maximum,
        "min": minimum
    }

# Test cases
list1 = [10, 20, 30, 40, 50]
print(f"Stats for {list1}: {stats(list1)}")

list2 = [5, 1, 9, 3, 7]
print(f"Stats for {list2}: {stats(list2)}")

empty_list = []
print(f"Stats for {empty_list}: {stats(empty_list)}")
```

```python
def calculator(num1, num2, operation):
    if operation == "+":
        return num1 + num2
    elif operation == "-":
        return num1 - num2
    elif operation == "*":
        return num1 * num2
    elif operation == "/":
        if num2 == 0:
            return "Error: Cannot divide by zero!"
        return num1 / num2
    else:
        return "Error: Invalid operation. Please use '+', '-', '*', or '/'."

# Example usage:
print(f"10 + 5 = {calculator(10, 5, '+')}")
print(f"10 - 5 = {calculator(10, 5, '-')}")
print(f"10 * 5 = {calculator(10, 5, '*')}")
print(f"10 / 5 = {calculator(10, 5, '/')}")
print(f"10 / 0 = {calculator(10, 0, '/')}")
print(f"10 ^ 5 = {calculator(10, 5, '^')}")
```

```python
def add(x, y):    return x + y
def subtract(x, y): return x - y
def multiply(x, y): return x * y
def divide(x, y):
    if y == 0: return "Cannot divide by zero!"
    return x / y

def calculator():
    print("=== Smart Calculator ===")
    while True:
        print("\n1. Add\n2. Subtract\n3. Multiply\n4. Divide\n5. Exit")
        choice = input("Choose (1-5): ")

        if choice == "5":
            print("Goodbye!")
            break

        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        if choice == "1":
            print("Result =", add(num1, num2))
        elif choice == "2":
            print("Result =", subtract(num1, num2))
        elif choice == "3":
            print("Result =", multiply(num1, num2))
        elif choice == "4":
            print("Result =", divide(num1, num2))
        else:
            print("Invalid choice!")

calculator()
```

```python
import csv

def save_data(filename, data):
    """
    Saves a list of lists as a CSV file.

    Args:
        filename (str): The name of the CSV file to create.
        data (list of list): A list of lists representing the data to save.
                             Each inner list is a row in the CSV.
    """
    try:
        with open(filename, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(data)
        print(f"Data successfully saved to '{filename}'.")
    except IOError as e:
        print(f"Error saving data to file: {e}")

# Example usage:
example_data = [
    ["Name", "Age", "City"],
    ["Alice", 30, "New York"],
    ["Bob", 24, "London"],
    ["Charlie", 35, "Paris"]
]

save_data("example.csv", example_data)

# You can also verify the content by reading it back
print("\nVerifying content of example.csv:")
with open("example.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
#Function that reads CSV and returns list of dict
import csv

def read_csv_as_dict(filename):
    """
    Reads a CSV file and returns its content as a list of dictionaries.
    Each dictionary represents a row, with column headers as keys.

    Args:
        filename (str): The name of the CSV file to read.

    Returns:
        list: A list of dictionaries, where each dictionary is a row.
              Returns an empty list if the file is not found or empty.
    """
    data_list = []
    try:
        with open(filename, 'r', newline='') as file:
            reader = csv.DictReader(file)
            for row in reader:
                data_list.append(row)
        print(f"Data successfully read from '{filename}'.")
        return data_list
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
        return []
    except Exception as e:
        print(f"An error occurred: {e}")
        return []

# Example usage with the previously created 'example.csv':
data_from_csv = read_csv_as_dict("example.csv")

if data_from_csv:
    print("\nContent as list of dictionaries:")
    for row_dict in data_from_csv:
        print(row_dict)
```

```python
def add_task(task_description):
    """Appends a new task to the todo.txt file."""
    with open("todo.txt", "a") as file:
        file.write(task_description + "\n")
    print(f"Task '{task_description}' added.")

def view_tasks():
    """Reads and prints all tasks from the todo.txt file."""
    try:
        with open("todo.txt", "r") as file:
            tasks = file.readlines()
        if not tasks:
            print("Your todo list is empty!")
        else:
            print("\n--- Your Todo List ---")
            for i, task in enumerate(tasks, 1):
                print(f"{i}. {task.strip()}")
            print("----------------------")
    except FileNotFoundError:
        print("No todo list found. Start by adding a task!")

def todo_manager():
    """Main function to manage the todo list."""
    print("=== Todo List Manager ===")
    while True:
        print("\n1. Add Task\n2. View Tasks\n3. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            task = input("Enter the task description: ")
            add_task(task)
        elif choice == '2':
            view_tasks()
        elif choice == '3':
            print("Exiting Todo List Manager. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

# Run the todo manager
todo_manager()
```

```python
#Day 7 Object-oriented Programming (OOP) in python
# ================== CLASS & OBJECT ==================
class Person:
    def __init__(self, name, age):   # Constructor
        self.name = name             # Instance attributes
        self.age = age

    def introduce(self):             # Method
        print(f"Hi, I'm {self.name} and I'm {self.age} years old.")

    def have_birthday(self):
        self.age += 1
        print(f"Happy Birthday {self.name}! Now you are {self.age}.")

# Creating objects
p1 = Person("Alex", 21)
p2 = Person("Sara", 19)

p1.introduce()
p2.introduce()
p1.have_birthday()
```

```python
# ================== INHERITANCE ==================
class Student(Person):               # Student inherits from Person
    def __init__(self, name, age, grade, subjects):
        super().__init__(name, age)  # Call parent constructor
        self.grade = grade
        self.subjects = subjects

    def study(self):
        print(f"{self.name} is studying {self.subjects}")

class Teacher(Person):
    def __init__(self, name, age, subject, salary):
        super().__init__(name, age)
        self.subject = subject
        self.salary = salary

    def teach(self):
        print(f"{self.name} is teaching {self.subject}")

# Using inheritance
s1 = Student("Emma", 20, "A", ["Math", "Python"])
t1 = Teacher("Mr. Sharma", 45, "AI", 75000)

s1.introduce()
s1.study()
t1.introduce()
t1.teach()
```

```python
# ================== ENCAPSULATION (Private attributes) ==================
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance   # private (name mangling)

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}. New balance: {self.__balance}")

    def get_balance(self):
        return self.__balance

acc = BankAccount("Alex", 5000)
acc.deposit(2000)
print(acc.get_balance())
# print(acc.__balance)   # This will raise error
```

```python
#Exercise 11: class account with a withdraw method and balance check
class BankAccount():
  def __init__(self, owner, initial_balance=0):
    self.owner = owner
    self.__balance = initial_balance # Make balance private

  def deposit(self, amount):
    if amount > 0:
      self.__balance += amount
      print(f"Deposited ${amount:.2f}. New balance: ${self.__balance:.2f}")
    else:
      print("Deposit amount must be positive.")

  def withdraw(self, amount):
    if amount > 0:
      if self.__balance >= amount:
        self.__balance -= amount
        print(f"Withdrew ${amount:.2f}. New balance: ${self.__balance:.2f}")
      else:
        print(f"Insufficient balance. Current balance: ${self.__balance:.2f}. Attempted withdrawal: ${amount:.2f}")
    else:
      print("Withdrawal amount must be positive.")

  def get_balance(self):
    return self.__balance

# Create an object and test the methods
my_account = BankAccount("Alice", 1000)
print(f"Initial balance for {my_account.owner}: ${my_account.get_balance():.2f}")

my_account.deposit(500)
my_account.withdraw(200)
my_account.withdraw(1500) # This should fail
my_account.deposit(200)
my_account.withdraw(1500) # This should now pass
print(f"Final balance for {my_account.owner}: ${my_account.get_balance():.2f}")
```

```python
#Exercise 12; student mark, average and pass status
class Student:
  def __init__(self, name, subject1, subject2, subject3):
    self.name = name
    self.subject1 = subject1
    self.subject2 = subject2
    self.subject3 = subject3

  def calculate_average(self):
    return (self.subject1 + self.subject2 + self.subject3) / 3

  def is_pass(self):
    average = self.calculate_average()
    if average >= 40:
      return True
    else:
      return False

#create object
s1 = Student("Samuel", 66, 49, 87)
s2 = Student("Funmi", 34, 44, 67)
s3 = Student("Temitope", 55, 48, 73)

print(f"The average score of {s1.name} is {s1.calculate_average():.2f}. Pass status: {s1.is_pass()}")
print(f"The average score of {s2.name} is {s2.calculate_average():.2f}. Pass status: {s2.is_pass()}")
print(f"The average score of {s3.name} is {s3.calculate_average():.2f}. Pass status: {s3.is_pass()}")
```

```python
#Exercise 15: create class library and add books as object
class Library:
  def __init__(self, name):
    self.name = name
    self.books = []  # Initialize an instance variable for books

  def add_book(self, book):
    self.books.append(book)
    print(f"'{book}' has been added to {self.name}.")

  def display_books(self):
    if not self.books:
      print(f"{self.name} is currently empty.")
    else:
      print(f"\nBooks in {self.name}:")
      for i, book in enumerate(self.books, 1):
        print(f"{i}. {book}")

# Create objects
l1 = Library("My Library")
l2 = Library("Your Library")

l1.add_book("The Great Gatsby")
l1.add_book("1984")

l2.add_book("To Kill a Mockingbird")
l2.add_book("Brave New World")

l1.display_books()
l2.display_books()
```

```python
#Mini project: student and Employee management system
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"Hi, I'm {self.name}, {self.age} years old.")

class Student(Person):
    def __init__(self, name, age, roll_no, marks):
        super().__init__(name, age)
        self.roll_no = roll_no
        self.marks = marks  # dictionary of subjects

    def get_average(self):
        return sum(self.marks.values()) / len(self.marks)

    def display(self):
        print(f"Student: {self.name} | Roll: {self.roll_no} | Avg: {self.get_average():.2f}")

class Employee(Person):
    def __init__(self, name, age, emp_id, salary, department):
        super().__init__(name, age)
        self.emp_id = emp_id
        self.salary = salary
        self.department = department

    def display(self):
        print(f"Employee: {self.name} | ID: {self.emp_id} | Dept: {self.department} | Salary: {self.salary}")

# Using the system
students = []
employees = []

# Add sample data
s1 = Student("Emma", 20, "S101", {"Math": 85, "Python": 92})
e1 = Employee("Mr. Rao", 35, "E201", 65000, "AI")

students.append(s1)
employees.append(e1)

print("=== MANAGEMENT SYSTEM ===")
print("Students:")
for s in students:
    s.display()

print("\nEmployees:")
for e in employees:
    e.display()
```

```python
#Exercise 15
import numpy as np

def array_stats(arr):
    """
    Takes a NumPy array and returns a dictionary with basic statistics.
    """
    return {
        "sum": np.sum(arr),
        "mean": np.mean(arr),
        "max": np.max(arr),
        "min": np.min(arr),
        "shape": arr.shape
    }

# Test the function
test_arr = np.random.rand(3, 4)
stats = array_stats(test_arr)
print("Array Statistics:")
for key, value in stats.items():
    print(f"{key.capitalize()}: {value}")
```

```python
import numpy as np

def matrix_calculator():
    print("=== NumPy Matrix Calculator ===")

    # Create two matrices
    print("Enter elements for Matrix A (2x2):")
    a = np.array([[int(input()), int(input())],
                  [int(input()), int(input())]])

    print("Enter elements for Matrix B (2x2):")
    b = np.array([[int(input()), int(input())],
                  [int(input()), int(input())]])

    print("\nMatrix A:\n", a)
    print("Matrix B:\n", b)

    print("\n1. Addition\n2. Subtraction\n3. Element-wise Multiplication\n4. Matrix Multiplication\n5. Exit")
    choice = int(input("Choose operation: "))

    if choice == 1:
        print("A + B =\n", a + b)
    elif choice == 2:
        print("A - B =\n", a - b)
    elif choice == 3:
        print("A * B =\n", a * b)
    elif choice == 4:
        print("A @ B =\n", a @ b)
    elif choice == 5:
        print("Goodbye!")
    else:
        print("Invalid choice")

matrix_calculator()
```
